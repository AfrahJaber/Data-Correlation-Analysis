`Correlations` : is a statistical measure that expresses the extent to which two variables are linearly related.


We can find Correlations from the Features (Columns) in this dataset using heat map from seaborne lib. Let's see how it works:


### Do you think the parent Education might effect in the students score?
# Let's see :)

```python
# create a new df we can call it df1 to make a groupby bettwen ParentEduc and the mean of the score of each test like this :
# ÙƒÙŠÙ ÙŠØ¤Ø«Ø± ØªØ¹Ù„ÙŠÙ… Ø§Ù„ÙˆØ§Ù„Ø¯ÙŠÙ† Ø¹Ù„Ù‰ Ø¯Ø±Ø¬Ø§Øª Ø§Ù„Ø·Ø§Ù„Ø¨
df1=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'} , inplace = True)
print(df1)
```

```python
# now we can do the heatmap :

sns.heatmap(df1 , annot= True)
plt.title("how parent's Education effect on student's score")
plt.show()
```

**Now we can see the correlations bettwen the parents who have master degree the students Scores.**

**So we can findout : That the student's who have High scores there parents have master dgree.**



#**Do you think that test preparation may affect students' grades?.**

```python
df2=df.groupby("TestPrep").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'} , inplace = True)
print(df2)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(df2 , annot= True)
plt.title("How does studying before an exam affect a student's grades?")
plt.show()
```

#### Now we can see that the students who attended the test got higher scores than the others.

####So we can find out: the students who got high scores are the ones who attended the test beforehand.

####Does university affiliation have an impact on academic performance?

```python
df3=df.groupby("EthnicGroup").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'} , inplace = True)
print(df3)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(df3 , annot= True)
plt.title("Does university affiliation have an impact on academic performance?")
plt.show()
```

#**Are there differences in academic performance between males and females in mathematics, reading, and writing?**

```python
df4=df.groupby("Gender").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'} , inplace = True)
print(df4)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(df4 , annot= True)
plt.title("differences in academic performance between males and females")
plt.show()
```
